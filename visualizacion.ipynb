# Cuaderno para la visualizaci√≥n del dataframe imdb.csv

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("imdb_clean.csv")
```

```python
df["averageRating"].plot.kde()
```

```python
df["averageRating"].plot.box()
```

```python
df["genres"].mode()
```

```python
animated_shows = df[df["Animated"] == "yes"]
```

```python
votes_per_year = animated_shows.groupby("releaseYear")["numVotes"].sum().reset_index()
```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(data=votes_per_year, x='releaseYear', y='numVotes', marker='o')
plt.title('Total Votes for Animated TV Shows by Year')
plt.xlabel('Release Year')
plt.ylabel('Total Number of Votes')
plt.grid(True)
plt.show()
```

```python
numerical_columns = [col for col in df.columns if df[col].dtype != "O"]
correlation_matrix = df[numerical_columns].corr()
correlation_matrix = correlation_matrix.corr()

# Plotting the heatmap using seaborn
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", linewidths=0.5)
plt.title('Correlation Matrix of releaseYear, numVotes, and averageRating')
plt.show()
```
